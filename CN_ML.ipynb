Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_openml
from sklearn import datasets
boston = fetch_openml(name='boston', version=1)
x = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.Series(boston.target, name='MEDV')
x
```

```python
df=x.copy()
df["CRIM2"]=df["CRIM"]**2
df["INDUS2"]=df["INDUS"]**2
df["NOX2"]=df["NOX"]**2
df["RM2"]=df["RM"]**2
df["AGE2"]=df["AGE"]**2
df["DIS2"]=df["DIS"]**2
df["TAX2"]=df["TAX"]**2
df["PTRATIO2"]=df["PTRATIO"]**2
df["B2"]=df["B"]**2
df["LSTAT2"]=df["LSTAT"]**2
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train1, x_test1, y_train1, y_test1 = train_test_split(x, y, test_size=0.25, random_state=42)
x_train2, x_test2, y_train2, y_test2 = train_test_split(df.values, y, test_size=0.25, random_state=42)
lg1=LinearRegression()
lg2=LinearRegression()
lg1.fit(x_train1, y_train1)
lg2.fit(x_train2, y_train2)
```

```python
x_train1 = np.array(x_train1)
y_train1 = np.array(y_train1)
x_test1 = np.array(x_test1)
y_test1 = np.array(y_test1)
x_train2 = np.array(x_train2)
y_train2 = np.array(y_train2)
x_test2 = np.array(x_test2)
y_test2 = np.array(y_test2)

score1=lg1.score(x_train1, y_train1)
score2=lg2.score(x_train2, y_train2)
score11=lg1.score(x_test1, y_test1)
score22=lg2.score(x_test2, y_test2)
print("Training score 1: ",score1)
print("Testing score 1: ",score11)
print("Training score 2: ",score2)
print("Testing score 2: ",score22)
```

Generic Gradient Descent

```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
import numpy as np

boston = fetch_openml(name='boston', version=1)
x = np.array(boston.data, dtype=float)  # Convert data to numpy array
y = np.array(boston.target, dtype=float)  # Convert target to numpy array and float
ones_column = np.ones((x.shape[0], 1))  # len(x.row) as row length and 1 as column length
x = np.hstack((x, ones_column))

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

def cal_mc(x, y, m, alpha):
    m_deriv = np.zeros(len(m))  # Initialize derivative vector

    for i in range(len(x)):
        prediction = np.dot(m, x[i])  # Calculate the prediction using dot product
        for j in range(len(m)):       # Calculate the derivative for each feature
            m_deriv[j] += (-2/len(x)) * (y[i] - prediction) * x[i][j]
    m -= alpha * m_deriv  # Update m using gradient descent step
    return m

def cal_cost(x, y, m):
    cost = 0
    for i in range(len(x)):
        prediction = np.dot(m, x[i])
        cost += (y[i] - prediction) ** 2
    return cost

def gd(x, y, alpha, itr):
    m = np.zeros(x.shape[1])  # Initialize m with zeros, assuming x.shape[1] is number of features

    for i in range(itr):
        m = cal_mc(x, y, m, alpha)
        cost = cal_cost(x, y, m)
        print("Iteration:", i, " Cost:", cost)

    return m

m = gd(x_train, y_train, 0.0000032, 1000)
print("Final coefficients:", m)
```

```python
y_train_pred=np.dot(x_train,m)
y_test_pred=np.dot(x_test,m)
```

```python
def score(y_true, y_pred):
    u = ((y_true - y_pred) ** 2).sum()
    v = ((y_true - y_true.mean()) ** 2).sum()
    return 1 - u / v
```

```python
score1=score(y_train,y_train_pred)
score2=score(y_test,y_test_pred)
print("Training score 1: ",score1)
print("Testing score 1: ",score2)
```
